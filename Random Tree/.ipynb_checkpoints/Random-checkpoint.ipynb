```python
import pandas as pd
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
dir(digits)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.gray()
for i in range(5):
    plt.matshow(digits.images[i])
```

```python
digits.data[:5]
```

```python
df = pd.DataFrame(digits.data)  # by using dataframe we can get neet datas
df
```

```python
df['target'] =digits.target
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df.drop(['target'],axis='columns'),digits.target,test_size = 0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier     # ensemble used while multiple algorithms take place
model = RandomForestClassifier(n_estimators = 10)    # n_estimators will tune the probability
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
y_predicted = model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_predicted)
cm
```

```python
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot = True)
plt.xlabel('predicted')
plt.ylabel('Truth')
```
